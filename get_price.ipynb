```python
import efinance as ef
import time
from datetime import datetime, timedelta

stock_code = 'SPX'
freq = 1
status = {stock_code: 0}

while True:
    df = ef.stock.get_quote_history(
        stock_code, klt=freq)

    now = str(datetime.today()).split('.')[0][:-3]
    print(now)

    founded = False
    for i in range(len(df)):
        if df["日期"][i] == now:
            print(df["收盘"][i])
            founded = True
            trend = 0
            if df["收盘"][i] > df["收盘"][i-1]:
                trend = 1
            if df["收盘"][i] < df["收盘"][i-1]:
                trend -1
            with open("trend.txt", "a") as f:
                f.write("{},{}\n".format(now, trend))

    if not founded:
        print("all collected")
        break

    print(f'saved {now}, stock: {stock_code} info into: trend.txt ！')

    print("pause for 1 minute")
    time.sleep(60)
    


# print('all collected')
```
